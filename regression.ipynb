```python
# lets import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler , OneHotEncoder
```

```python
# lets bring dat to field
df=pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df=df.drop('Id', axis=1)
```

```python
df.isna().sum()
```

```python
# lets create the train and test spilt
x=df.drop(['SalePrice'], axis=1)
y=df['SalePrice']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# lets clarify all those columns even they are categoricak l or numerical
num_cols=[]
cat_cols=[]

for col in x_train.columns:
  if x_train[col].dtype=='object':
    cat_cols.append(col)
  else:
    num_cols.append(col)
```

```python
num_cols, cat_cols
```

```python
# lets create make transformer

ct=make_column_transformer(
    (MinMaxScaler(), ['MSSubClass',
  'LotArea',
  'OverallCond',
  'YearBuilt',
  'YearRemodAdd',
  'BsmtFinSF2',
  'TotalBsmtSF']),
    (OneHotEncoder(handle_unknown='ignore'),['MSZoning', 'LotConfig', 'BldgType', 'Exterior1st'] )
)

# Convert categorical columns to string type
for col in cat_cols:
    x_train[col] = x_train[col].astype(str)
    x_test[col] = x_test[col].astype(str)

# lets fit on train data
ct.fit(x_train)
# lets transform on test and train

x_train_normal=ct.transform(x_train)
x_test_normal=ct.transform(x_test)
```

```python
# now lets create the first model to train
tf.random.set_seed(42)
model_reg= tf.keras.models.Sequential([
    tf.keras.layers.Dense(4),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='linear')




])

# lets copile the model
model_reg.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.Adam(.01),
    metrics=['mae']
)


model_reg.fit(x_train_normal, y_train,
              validation_data=(x_test_normal, y_test),
              epochs=20)
```
